```python
def run_adam(model, iterations, data):
    """
    Utility function running the Adam optimizer

    :param model: GPflow model
    :param interations: number of iterations
    """
    # Create an Adam Optimizer action
    logf = []
    training_loss = model.training_loss_closure(data)
    optimizer = tf.optimizers.Adam()

    @tf.function
    def optimization_step():
        optimizer.minimize(training_loss, model.trainable_variables)

    for step in range(iterations):
        optimization_step()
        if step % 10 == 0:
            elbo = -training_loss().numpy()
            logf.append(elbo)
            clear_output(wait=True)
            print("iteration {}".format(step))
    return logf
```

```python
def training_loop(
    model,
    data,
    optimizer = None,
    maxiter=1e3,
    compile=False,
):
    """
    Simple generic training loop. At each iteration uses a GradientTape to compute
    the gradients of a loss function with respect to a set of variables.

    :param closure: Callable that constructs a loss function based on data and model being trained
    :param optimizer: tf.optimizers or tf.keras.optimizers that updates variables by applying the
        corresponding loss gradients. Adam is a default optimizer with default settings.
    :param var_list: List of model variables to be learnt during training
    :param maxiter: Maximum number of
    :return:
    """

    optimizer = tf.optimizers.Adam() if optimizer is None else optimizer
    var_list: List[tf.Variable] = model.trainable_variables
    def optimization_step():
        with tf.GradientTape(watch_accessed_variables=False) as tape:
            tape.watch(var_list)
            loss = -elbo(model,data)
            print(loss)
        grads = tape.gradient(loss, var_list)
        optimizer.apply_gradients(zip(grads, var_list))

    if compile:
        optimization_step = tf.function(optimization_step)

    for _ in range(int(maxiter)):
        optimization_step()
```

```python
def get_Kuu(model, Znew, Zold):
    
    K = model.kernel
    
    Kxx = K(Znew)
    Kxm = K(Znew, Zold)
    Kmm = K(Zold, Zold)
    Kmminv = tf.linalg.inv(Kmm)
    #Kmx = K(Zold,Znew)
    Kmx = tf.transpose(Kxm)
    phi_L = m_old.q_sqrt
    A = np.empty((1, 3, 3))
    [np.dot(phi_L[q, :, :], tf.transpose(phi_L[q, :, :]), A[q, :, :]) for q in range(1)]
    #A = model.q_sqrt[0,:,:]
    B = Kmminv @ A @ Kmminv
    
    Kuu = (
            Kxx - 
            (Kxm @ Kmminv @ Kmx) +
            (Kxm @ B @ Kmx)
            )
    
    mu_var = Kxz @ Kmminv @ model.q_mu
    
    return Kuu, mu_var
```

```python
from GPy.util import choleskies
from GPy.util import linalg
import tensorflow as tf

def conditional_prior(Z, Zold, kern_list_old, phi_means, phi_chols):
    M, Dz = Z.shape
    Mold, _ = Zold.shape
    Q = len(kern_list_old)

    # Algebra for q(u):
    #phi_m = phi_means.copy()
    phi_L = phi_chols
    phi_S = np.empty((Q, Mold, Mold))
    [np.dot(phi_L[q, :, :], tf.transpose(phi_L[q, :, :]), phi_S[q, :, :]) for q in range(Q)]

    Mu = np.empty((Q, M, 1))
    Kuu = np.empty((Q, M, M))
    Luu = np.empty((Q, M, M))
    Kuui = np.empty((Q, M, M))

    Kuu_old, Luu_old, Kuui_old = latent_funs_cov(Zold, kern_list_old)
    Kuu_new, _, _ = latent_funs_cov(Z, kern_list_old)
    Kuu_cond = latent_funs_conditional(Z, Zold, kern_list_old)

    for q, kern in enumerate(kern_list_old):
        R, _ = linalg.dpotrs(np.asfortranarray(Luu_old[q, :, :]), Kuu_cond[q, :, :].T)
        Auu = R.T # Kuu_cond * Kuui
        Mu[q, :] = np.dot(Auu, phi_means[:, q, None])
        Kuu[q, :, :] = Kuu_new[q, :, :] + np.dot(np.dot(R.T, phi_S[q,:,:]), R) - np.dot(Kuu_cond[q, :, :], R)
        Luu[q, :, :] = linalg.jitchol(Kuu[q, :, :])
        Kuui[q, :, :], _ = linalg.dpotri(np.asfortranarray(Luu[q, :, :]))
    return Mu, Kuu, Luu, Kuui, phi_S

def latent_funs_cov(Z, kernel_list):
    """
    Builds the full-covariance cov[u(z),u(z)] of a Multi-output GP
    for a Sparse approximation
    :param Z: Inducing Points
    :param kernel_list: Kernels of u_q functions priors
    :return: Kuu
    """
    Q = len(kernel_list)
    M,Dz = Z.shape
    Xdim = int(Dz/Q)
    #Kuu = np.zeros([Q*M,Q*M])
    Kuu = np.empty((Q, M, M))
    Luu = np.empty((Q, M, M))
    Kuui = np.empty((Q, M, M))
    for q, kern in enumerate(kernel_list):
        Kuu[q, :, :] = kern.K(Z[:,q*Xdim:q*Xdim+Xdim],Z[:,q*Xdim:q*Xdim+Xdim])
        Luu[q, :, :] = linalg.jitchol(Kuu[q, :, :])
        Kuui[q, :, :], _ = linalg.dpotri(np.asfortranarray(Luu[q, :, :]))
    return Kuu, Luu, Kuui

def latent_funs_conditional(Z, Zold, kernel_list):
    Q = len(kernel_list)
    M, Dz = Z.shape
    Mold, Dz_old = Zold.shape
    Xdim = int(Dz/Q)
    Kuu_cond = np.empty((Q, M, Mold))
    for q, kern in enumerate(kernel_list):
        Kuu_cond[q, :, :] = kern.K(Z[:,q*Xdim:q*Xdim+Xdim], Zold[:,q*Xdim:q*Xdim+Xdim])
    return Kuu_cond
```

```python
class ContinuousGP(gpflow.kernels.Kernel):
    def __init__(self, model, Z):
        self.old_Z = Z
        self.old_model = model
        super().__init__(active_dims=[0])
        self.variance = gpflow.utilities.deepcopy(model.kernel.variance)
        self.lengthscales = gpflow.utilities.deepcopy(model.kernel.lengthscales)
        gpflow.set_trainable(self,True)
        gpflow.set_trainable(self.old_model,False)
        
        
    def k_ss(self, X, X2):
        m = self.old_model
        _, cov = m.predict_f(tf.concat((X,X2),0), full_cov=True)
        postcov = cov[0,:X.shape[0],X.shape[0]:]
        return postcov

    def k_su(self, X):
        m = self.old_model
        Z = self.old_Z
        _, cov = m.predict_f(tf.concat((X,Z),0), full_cov=True)
        postcov = cov[0,:X.shape[0],X.shape[0]:]
        return postcov
    
    def k_us(self, X2):
        m = self.old_model
        Z = self.old_Z
        _, cov = m.predict_f(tf.concat((X2,Z),0), full_cov=True)
        postcov = cov[0,:X2.shape[0],X2.shape[0]:]
        return postcov

    def K_uu(self):
        m = self.old_model
        Z = self.old_Z
        _, cov = m.predict_f(tf.concat((Z,Z),0), full_cov=True)
        postcov = cov[0,:Z.shape[0],Z.shape[0]:]
        return postcov
    
    def K(self, X, X2=None):
        
        m = self.old_model
                
        if X2 is None:
            X2 = X
            
        k_ss = self.k_ss(X,X2)
        k_su = self.k_su(X)
        k_us = tf.transpose(self.k_us(X2))
        K_uu = self.K_uu()
        K_uu_inv = tf.linalg.inv(K_uu)
        
        A = m.q_sqrt[0,:,:]
        B = K_uu_inv @ A @ K_uu_inv
        
        M1 = k_ss
        M2 = k_su @ K_uu_inv  @ k_us
        M3 = k_su @ B @ k_us       
       
        postcov = tf.add(k_ss, (k_su @ K_uu_inv) @ tf.subtract(A, K_uu) @ (K_uu_inv @ k_us))    
                
        return postcov

    def K_diag(self, X):
        return tf.linalg.diag_part(self.K(X)) # self.variance * tf.reshape(X, (-1,))  # this returns a 1D tensor
        #np.diag(self.K(X,X)) #lazy #
```

```python
class ContinuousMean(gpflow.mean_functions.MeanFunction):
    def __init__(self, model, kernel, Z):
        self.model = model
        self.K = kernel
        self.Z = Z
    def __call__(self,X):
        kernel = self.K
        Z = self.Z
        return kernel.k_su(X,Z) @ kernel.K_uu_inv(Z) @ kernel.self.model.q_mu
```
