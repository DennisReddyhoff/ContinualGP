```python
%matplotlib inline
import itertools
import numpy as np
import time
import gpflow
import tensorflow as tf
import matplotlib.pyplot as plt
from gpflow.ci_utils import ci_niter
from gpflow.utilities import print_summary, set_trainable, deepcopy
from IPython.display import clear_output

# for reproducibility of this notebook:
tf.random.set_seed(42)

N = 1000 # number of samples
M = 3  # number of inducing points
Q = 1  # number of latent function
T = 10 # number of streaming batches
max_X = 0.4 # max of X range
VEM_its = 2
max_iter = 100
```

```python
X = np.sort(max_X*np.random.rand(N))[:, None]
def true_f(X_input):
    return (4.5 * np.cos(2 * np.pi * X_input + 1.5*np.pi) - \
           3 * np.sin(4.3 * np.pi * X_input + 0.3 * np.pi) + \
           5 * np.cos(7 * np.pi * X_input + 2.4 * np.pi))    

Y = np.random.normal(true_f(X))
data = (X,Y)
```

```python
plt.plot(X,Y)
Xt = np.linspace(0, 0.4, 1000)[:, None]
Yt = true_f(Xt)
_ = plt.plot(Xt, Yt, c="k")
```

```python
m = gpflow.models.GPR((X, Y),kernel=gpflow.kernels.RBF(1.0))
```

```python
m.kernel
```

```python
def run_adam(model, N):
    optimizer=tf.optimizers.Adam()
    for i in range(N):
        optimizer.minimize(model.training_loss,model.trainable_variables)    
        if i % 10 == 0:        
            clear_output(wait=True)
            print("iteration {}".format(i))
            
```

```python
run_adam(m,500)
```

```python
plt.plot(X,Y)
Xt = np.linspace(0, 0.4, 1000)[:, None]
Yt, Yv  = m.predict_f(Xt)
plt.plot(Xt, Yt, c="k")
plt.fill_between(
        Xt[:, 0],
        (Yt - 2 * Yv ** 0.5)[:, 0],
        (Yt + 2 * Yv ** 0.5)[:, 0],        
        alpha=1,
        lw=1.5,
    )
```

```python
class ContinuousGP(gpflow.kernels.Kernel):
    def __init__(self, model):
        super().__init__(active_dims=[0])
        self.variance = gpflow.utilities.deepcopy(model.kernel.variance)
        self.lengthscales = gpflow.utilities.deepcopy(model.kernel.lengthscales)
    
    def K(self, X, X2=None):
        if X2 is None:
            X2 = X
        _, cov = m_old.predict_f(np.r_[X,X2], full_cov=True)
        postcov = cov[0,:X.shape[0],X.shape[0]:]
        return postcov

    def K_diag(self, X):
        return tf.linalg.diag_part(self.K(X)) # self.variance * tf.reshape(X, (-1,))  # this returns a 1D tensor
        #np.diag(self.K(X,X)) #lazy #

k_new = ContinuousGP(m_old) 
```
